```python
# import numpy as np
import matplotlib.pyplot as plt
# import pandas as pd
```

**We will be using a small Data for demonstrating linear regression

```python
weights = [50, 60, 70, 80, 90, 100]  # Independent variable (X)
heights = [150, 160, 170, 175, 180, 185]  # Dependent variable (Y)
```

```python
x = 0
for i in weights:
    x += i
mean_weight_x  = x / len(weights) # This is 75


y = 0
for k in heights:
    y += k
mean_height_y = y / len(heights) # This is 170
```

**Deviations of Weights (X) , Heights(Y) and Sum of square of deviations**

```python
deviations_x  = []
for i in range(len(weights)):
    deviations_x.append(weights[i] - mean_weight_x)
# print(deviations_x)  [-25.0, -15.0, -5.0, 5.0, 15.0, 25.0]



deviations_y = []
for i in range(len(heights)):
    deviations_y.append(heights[i] - mean_height_y)
# print(deviations_y)  [-20.0, -10.0, 0.0, 5.0, 10.0, 15.0]


deviations_product = []
for i in range(len(deviations_x)):
    deviations_product.append(deviations_x[i] * deviations_y[i])
# deviations_product  [500.0, 150.0, -0.0, 25.0, 150.0, 375.0]


sum_deviations_product = 0
for x in deviations_product:
    sum_deviations_product += x
# print(sum_deviations_product)   This is 1200.0
 
square_deviations_x = [x**2 for x in deviations_x]
# square_deviations_x  [625.0, 225.0, 25.0, 25.0, 225.0, 625.0]

sum_square_deviations_x = 0
for x in square_deviations_x:
    sum_square_deviations_x += x
# print(sum_square_deviations_x)  1750.0
```

```python
slope = sum_deviations_product / sum_square_deviations_x
# slope  0.6857142857142857 
```

```python
b = mean_height_y - (slope * mean_weight_x)
b
```

```python
user_input_x =int(input("Enter Weight"))
prediction_height = slope*user_input_x + b  #y = mx + b
prediction_height
```

***Plotting Regression Line with current data and user input data***

```python
plt.scatter(weights, heights, color='blue', label="Data Points")  # Original data points
plt.plot(weights, [slope * x + b for x in weights], color='green', label="Regression Line")  # Regression line
plt.scatter(user_input_x, prediction_height, color='red', label=f"Predicted Point ({user_input_x}, {prediction_height:.2f})")  # Predicted point

# Labels and title
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Weight vs Height with Regression Line")
plt.legend()
plt.show()
```
